```python
from stackup_python import stackup
```

```python
help(stackup)
```

```python
size= 100000
dim1 = stackup.dimension(mean=30, tol=15, size=size)
dim2 = stackup.dimension(mean=40, tol=23, size=size)
```

```python
seed = 42
x = dim1.generate(seed)
y = dim2.generate(seed)
```

```python
x
```

```python
y
```

```python
import numpy as np
z = np.sqrt(x**2 + y**2)
z
```

```python
stackup.plot_dist(z)
```

```python
stackup.fit_dist(z)
```

```python
z.mean()
```

```python
z.std()
```

```python
3*z.std()
```

```python
z.mean() + 3*z.std()
```

```python
z.mean() - 3*z.std()
```

```python
z.min()
```

```python
z.max()
```
